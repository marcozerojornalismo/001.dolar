# A relação do dólar com a política brasileira

O código a seguir analisa a taxa de câmbio do dólar (compra) mês a mês, semestre a semestre e ano a ano, considerando apenas os últimos mandatos presidenciais (Dilma/Temer), iniciados em 2015.

A base de dados usada é do Banco Central, que pode ser baixada no seguinte link: http://dadosabertos.bcb.gov.br/dataset/10813-taxa-de-cambio---livre---dolar-americano-compra

O conteúdo foi baixado para a conversão de valores de 'objeto' para 'float' (substituição da ',' por '.').

Esta análise de dados é uma análise bruta e por isso não é completamente tratada. Serve de referência para trabalhos jornalísticos e de outras áreas sobre o tema.

A variação do dólar no período pode ter sido para mais ou para menos e se refere ao valor mínimo do período em comparação com o valor máximo.

O uso é livre para toda a população, desde que citada a fonte.
